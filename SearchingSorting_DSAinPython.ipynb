# **Sorting**

### `bubble_sort()`

```python
## GitHub copilot version
def bubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1): # last i elements are already in place (SORTED)
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j] #Swap
    return arr
```

```python
print(bubbleSort([5,4,3,2,1]))
```

```python
# My version
def BubbleSort(arr):
    for i in range(len(arr)):
        for j in range(i, len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i] #Swap
    return arr
## this is not correct bubble sort, but it is SOMEHOW WORKING
```

```python
arr = [1, 3, 4, 5, 2, 6, 9, 7, 8]
arr2 = [9, 8, 7, 6, 5, 4, 3, 2, 1]
print(BubbleSort(arr))
print(BubbleSort(arr2))
print("Time Complexity: O(n^2)") 
print("Space Complexity: O(1)")
```

```python
# Adaptive Bubble Sort 
# It is a technique to improve the Bubble Sort Algorithm. Knowledge of already sorted array is used.
def AdaBubbleSort(arr):
    for i in range(len(arr)):
        swapped = False
        for j in range(i, len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i] #Swap
                swapped = True
        if swapped == False:
            print("Already Sorted. break;")
            break

    return arr

arr = [1, 2, 3, 4, 5, 6, 7, 8, 9] # already sorted array
print(AdaBubbleSort(arr))
# O(n^2) in worst_case and O(n) in best_case 
```

### `selection_sort()`

```python
# GitHub copilot version
def selectionSort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i] #Swap
    return arr
```

```python
# My Version - without gpt helper
def SelectionSort(arr):
    for i in range(len(arr)):
        minIndex = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minIndex]:
                minIndex = j
        arr[i], arr[minIndex] = arr[minIndex], arr[i] #Swap
    return arr
```

```python
arr = [2, 5, 3, 1, 4, 6, 9, 7, 8]
arr2 = [9, 8, 7, 6, 5, 4, 3, 2, 1]
print(SelectionSort(arr))
print(SelectionSort(arr2))
print("Time Complexity: O(n^2)")
print("Space Complexity: O(1)")
```

### *Comparison* of `time_taken` by : **`BubbleSort() and SelectionSort()`** 

```python
import random
arr = [random.randint(100, 1000) for i in range(100000)]
print(arr)
print(len(arr))
```

```python
L1 = arr.copy()
L2 = arr.copy()

import time
start = time.time()
bubbleSort(L1)
end = time.time()
bubbleTime = end-start
print("Time taken by BubbleSort: ", end-start)

start = time.time()
selectionSort(L2)
end = time.time()
selectionTime = end-start
print("Time taken by SelectionSort: ", end-start)

print("SelectionSort is faster than BubbleSort by: ", bubbleTime-selectionTime)
```

### **``merge_sort()``**

```python
## GitHub copilot version
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]

        mergeSort(L)
        mergeSort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i+=1
            else:
                arr[k] = R[j]
                j+=1
            k+=1

        while i < len(L):
            arr[k] = L[i]
            i+=1
            k+=1

        while j < len(R):
            arr[k] = R[j]
            j+=1
            k+=1

    return arr
```

```python
def MergeSort(arr):
    print("Splitting ", arr)
    if len(arr) <= 1: # base case
        return arr
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:] 

    # divide and conquer approach -- continuosly deviding the array into smaller arrays of left and right then merging them in sorted manner later
    left = MergeSort(left) # recursive call
    right = MergeSort(right) # recursive call

    return merge_sorted(left, right) ## main sorting is kinda done here
```

```python
arr1 = [5, 9, 3, 1, 2, 8, 4, 7, 6]
print(MergeSort(arr1))
print()
print("Time Complexity: O(nlogn)")
print("Space Complexity: O(n)") # because of the auxillary array used in merge_sorted function to store the sorted array --> recusion stack space
```
